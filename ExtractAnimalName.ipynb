```python
from bs4 import BeautifulSoup
from urllib.request import urlopen

with open('animals.htm', 'r') as f:
    content = f.read()

soup = BeautifulSoup(content, 'html.parser')

links = soup.find_all('a')

animal_links = [link.get('href') for link in links if link.get('title') is not None and link.b is not None]
#print("animal_links=",animal_links)
animal_texts = list()

for animal_link in animal_links:
    print("Processing {}".format(animal_link))
    animal_soup = BeautifulSoup(urlopen(animal_link).read(), 'html.parser')
    for t in animal_soup.find_all('h2'):
        if t.get('class') is not None and t.get('class')[0] == 'az-article-title':
            title = t
    txt=title.findParent().text
    txt = txt[:txt.find('Share This Article')]
    txt = ' '.join(txt.split())
    txt = txt[txt.find('Location')+8:]  
    txt = txt[txt.find(title.text)+len(title.text):]
    clean_txt = txt[0]
    for k in range(1, len(txt)):
        if txt[k-1].isalpha() and txt[k].isalpha() and txt[k-1].islower() and txt[k].isupper():
            clean_txt += ' '
        clean_txt += txt[k]
    animal_texts.append(clean_txt)
    
with open('animal_texts_.txt', 'w') as f:
    f.write('\n\n'.join(animal_texts))
```
